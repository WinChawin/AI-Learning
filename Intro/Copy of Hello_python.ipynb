<h1> if

เป็นคำสั่งที่เอาไว้ตรวจสอบว่าการดำเนินการต่างๆเป็นจริงหรือไม่ ถ้าเป็นจริงจะเข้าเงื่อนไข แต่ถ้าไม่ก็จะปล่อยผ่านไป เช่น

```python
if 9 == 9:              # ถ้า 9 เท่ากับ 9 จะให้ทำบรรทัดถัดไปที่อยู่ข้างใน :
  print("เก้าเท่ากับเก้า")   # ถ้าเงื่อนไขเป็นจริงถึงจะทำคำสั่งนี้
```

!!!! จุดสำคัญคือต้องทำการ "เว้นวรรค" ให้ถูกต้อง เพราะระดับของโค้ดมีความสำคัญมากใน python โดยคำสั่งที่จะให้ทำหลังจากเข้าเงื่อนไข if จะต้องเว้นวรรค 1tab เสมอ

```python
if 9 == 8:              # ถ้า 9 เท่ากับ 8 จะให้ทำบรรทัดถัดไปที่อยู่ข้างใน :
  print("เก้าเท่ากับเก้า")   # คำสั่งนี้เว้นวรรคทำให้โปรแกรมรู้ว่าอยู่ภายใต้เงื่อนไขทางด้านบน ต้องรอเงื่อนไขเป็นจริงมันถึงจะทำ
print("อันนี้ไม่ได้อยู่ในif")  # คำสั่งนี้ไม่เว้นวรรคทำให้ต่อให้เงื่อนไขเป็นจริงหรือไม่มันก็ทำงานอยู่ดี
```

แล้วในกรณีที่ if ซ้อน if หล่ะ

```python
if 9==9:                 # ถ้า 9 เท่ากับ 9 ให้ทำคำสั่งภายใน
  if 9==8:               # เช็คอีกรอบว่าถ้า 9 เท่ากับ 8 ให้ทำข้างในอีกที
    print("เก้าเท่ากับแปด") # จะทำงานเมื่อ 9 เท่ากับ 9 และ 9 เท่ากับ 8
  print("เก้าเท่ากับเก้า")    # แค่ 9 เท่ากับ 9 ก็ทำงนแล้ว
```

<h1> elif

มีลักษณะเหมือน if แต่ใช้ในกรณีที่เงื่อนไขใน if ไม่เป็นจริง มันก็จะมาเช็ค elif ต่อ

```python
a = 20
if a <= 10:          #เงื่อนไขไม่เป็นจริง
  print('เข้าอันแรก')
elif a <= 20:        #เงื่อนไขเป็นจริง จึงทำอันนี้
  print('เข้าอันที่2')
elif a <= 30:        #เงื่อนไขเป็นจริงเหมือนกัน แต่โปรแกรมอ่านจากบนลงล่าง ถ้า elif บนเป็นจริงก่อน มันจะไม่เช็ค elif ตัวต่อไป 'โค้ดที่ใช่ต้องอยู่ให้ถูกที่ ไม่งั้นจะโดนโค้ดตัวบนคาบไปแดก'
  print('เข้าอันที่3')
```

<h1> else

ใช้สำหรับให้คำสั่งทำงานในกรณีที่ไม่เข้าเงื่อนไขทางด้านบนเลย

'ถ้าเขาไม่ใช่ เราพร้อมเป็นคนที่ใช่เสมอ'


```python
a = 20
if a == 10:
  print('เข้าอันแรก')
elif a == 15:
  print('เข้าอันที่2')
elif a == 30:
  print('เข้าอันที่3')
else: # จะสั่งให้คำสั่งใต้มันทำงานก็ต่อเมื่อไม่เข้าเงื่อนไขทางด้านบนเลย
  print('ไม่เข้าอะไรเลย')
```

<h1> or and

*   or จะเป็นจริงเมื่อมีเงื่อนไขเพียงหนึ่งอันเป็นจริง
*   and จะเป็นจริงเมื่อเงื่อนไขทั้งหมดเป็นจริง



```python
if (9==9 or 9==10):
  print('เงื่อนไขเป็นจริง')
else:
  print('เงื่อนไขไม่เป็นจริง')
# จะสังเกตุว่าแค่ 9เท่ากับ9 เป็นจริง โดยที่ตัวอื่นไม่ต้องเป็นจริง ก็เข้าเงื่อนไขแล้ว
```

```python
# โจทย์
# ให้สร้างฟังก์ชั่นในการตัดเกรดและฟังก์ชั่นในการคำนวนหาค่าbmi จากนั้นให้สร้างโปรแกรมในการเลือกเรียกใช้ฟังก์ชั่นเหล่านั้นอีกที
def menu(a, b , menu):
  if menu == 1:
    return bmi(a, b)
  elif menu == 2:
    return grade(a)
def bmi(w, h):
    return w / (h/100)**2

def grade(a):
    if a >= 80:
      return 4
    elif a >= 75:
      return 3.5
    elif a >= 70:
      return 3
    elif a >= 65:
      return 2.5
    elif a >= 60:
      return 2
    elif a >= 55:
      return 1.5
    elif a >= 50:
      return 1
    else:
      return 0
```

```python
def triangle(a, b, c):
  if a + b <= c:
    return"ไม่ได้"
  elif a + c <= b:
    return"ไม่ได้"
  elif c + b <= a:
    return"ไม่ได้"
  elif a == b == c:
    return"ด้านเท่า"
  elif a == b:
    return"หน้าจั่ว"
  elif a == c:
    return"หน้าจั่ว"
  elif b == c:
    return"หน้าจั่ว"
  elif a != b != c:
    return"ด้านไม่เท่า"

```

<H1>Nested Loop

พูดง่ายๆ มันคือลูปซ้อนลูปอะแหละ ตรงนี้ก็อาจจะทำให้หลายคนงงนิดนึง แต่เราจะพยายามนะ555

```python
for i in range(5):    # เริ่มแรกเราก็ให้ลูป5ครั้ง
  for j in range(5):  # แต่ละครั้งก็จะลูปไปอีก5รอบ
    print('*',end='') #  จะปริ้นดาวออกมา5ดวงทุกครั้งที่จบลูปใหญ่1รอบ
  print('')           # พอจบลูปใหญ่1รอบก็จะเว้นบรรทัด
```

```python
for i in range(5):    # เริ่มแรกเราก็ให้ลูป5ครั้ง
  for j in range(i+1):  # แต่ละครั้งก็จะลูปไปอีกตามค่าของi ซึ่งยิ่งรอบลูปมากขึ้น ดาวก็จะยิ่งมาก
    print('*',end='') # จะปริ้นดาวออกมาตามจำนวนรอบของลูปใหญ่
  print('')           # พอจบลูปใหญ่1รอบก็จะเว้นบรรทัด
```

```python
for i in range(5):    # เริ่มแรกเราก็ให้ลูป5ครั้ง
  for j in range(5-i):  # แต่ละครั้งก็จะลูปไปอีกตามค่าของจำนวนลูปใหญ่ทั้งหมดลบกับiที่เพิ่มขึ้นเรื่อยๆ ซึ่งหมายความว่าลูปเล็กก็จะน้อยลงเรื่อยๆ
    print('*',end='') # จะปริ้นดาวออกมาตามจำนวนรอบของลูปเต็มลบกับ i
  print('')           # พอจบลูปใหญ่1รอบก็จะเว้นบรรทัด
```

<h1>break continue

3ตัวช่วยที่ทำให้การกรองเงื่อนไขง่ายขึ้น เวลาใช้loop กับ if

```python
x = 1
while x <= 15:
  print(x)
  x = x +1
```

```python
### break คือคำสั่งหยุดลูป เมื่อเจอคำสั่งนี้ลูปจะหยุดทำงาน
while True:  # while loop เหมือนกับfor แต่ง่ายกว่า โดยถ้าเงื่อนไขหลังwhileเป็นจริงมันจะหมุนไปเรื่อยๆ ซึ่งtrueมันเท่ากับจริงอะนะ
  print('ทายซิว่าเลขนี้คืออ4ะไร')
  x = float(input())
  if x == 22: # ถ้าเลขเท่ากับ22
    break # สั่งจบลูปทันที
```

```python
# continue คือการผ่านลูปแบบไม่ต้องทำคำสั่งด้านล่างตัวมัน
for i in range(1,51): #ลูปเลขตั้งแต่ 1-50
  if i % 2 == 0: # ถ้าiหาร2ลงตัว
    continue # ถ้าโดนคำสั่งนี้มันจะผ่านไปขึ้นลูปใหม่โดยไม่อ่านบรรทัดด้านล่างเลย
  print(i)
```
